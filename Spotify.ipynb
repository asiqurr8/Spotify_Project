```python
#Scatter Plot

# Dataframe creation
df = pd.read_csv('/spotify.csv')

# Scatter plot creation
plt.figure(figsize=(12, 8))
plt.scatter(df.index, df['Rank'], color='green', s=60)

# Add titles and labels
plt.title('Scatter Plot of Song Rankings', fontsize=12)
plt.xlabel('Index', fontsize=16)
plt.ylabel('Rank', fontsize=16)

# Song and Title Annotation
for i, row in df.iterrows():
    plt.text(i, row['Rank'], f'{row["Song Title"]}\n{row["Artist"]}', fontsize=14, ha='center')

# Display plot
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(14, 8))
bars = plt.bar(df['Song Title'], df['Rank'], color='brown')

# Add title and labels
plt.title('Song Rankings')
plt.xlabel('Song Title')
plt.ylabel('Rank')

# We invert to show higher axis
plt.gca().invert_xaxis()

# Display plot
plt.xticks(rotation=90)
plt.show()
```

```python
# Matrix for each song
matrix = df.pivot(index='Song Title', columns='Rank', values='Rank')

# Heatmap plot
plt.figure(figsize=(14, 8))
sns.heatmap(matrix, annot=True, cmap='coolwarm', cbar=True, linewidths=1, linecolor='skyblue')

# Titles and labels adding
plt.title('Heatmap of Song Rankings', fontsize=16)
plt.xlabel('Rank', fontsize=14)
plt.ylabel('Song Title', fontsize=14)

# Display plot
plt.show()
```

```python
#Calculate mean, median, standard deviation, skewness, kurtosis
mean_rank = df['Rank'].mean()
median_rank = df['Rank'].median()
std_rank = df['Rank'].std()
data_skewness = df['Rank'].skew()
data_kurtosis = df['Rank'].kurtosis()

print(f"Mean : {mean_rank}")
print(f"Median : {median_rank}")
print(f"Standard Deviation : {std_rank}")
print(f"Skewness : {data_skewness}")
print(f"Kurtosis : {data_kurtosis}")
```

```python
#Numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Print the correlation matrix
print(corr_matrix)
```
